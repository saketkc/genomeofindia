# Plot zip codes in India

```r
suppressPackageStartupMessages({
  library(tidyverse)
  library(patchwork)
  library(ggpubr)
  library(data.table)
})
theme_set(theme_pubr())
```

```r
data <- setDF(fread("../data/IN//IN.txt"))  #, sep='\t', header=FALSE)
colnames(data) <- c("country_code", "postal_code", "place_name", "admin_name1", "admin_code1", "admiin_name2", "admin_code2", "admin_name3", "admin_code3",
  "latitude", "longitude", "accuracy")


data$starting_pin <- substr(as.character(data$postal_code), 1, 1)
data$starting_pin2 <- substr(as.character(data$postal_code), 2, 2)
data$starting_pin3 <- substr(as.character(data$postal_code), 3, 3)
data$starting_pin4 <- substr(as.character(data$postal_code), 4, 4)
data$starting_pin5 <- substr(as.character(data$postal_code), 5, 5)
data$starting_pin6 <- substr(as.character(data$postal_code), 6, 6)



```

```r
library(ggtext)
library(ggdark)
```

```r
sample(c("#191970", "#006400", "#ff0000", "#ffd700", "#00ff00", "#00ffff", "#ff00ff", "#ffb6c1"), size = 8)
```

```r
options(repr.plot.width=10, repr.plot.height=8)
ggplot(data, aes(longitude, latitude, color=starting_pin)) + geom_point(size=0.01) + 
scale_color_brewer(type="qual", palette="Set1", name = "1st digit of pincode") +
#scale_color_manual(values=sample(c("#191970","#006400","#ff0000","#ffd700","#00ff00","#00ffff","#ff00ff","#ffb6c1"), size=8))+
#hrbrthemes::theme_ipsum() + 
theme_pubr() +
  # xlab('\nDate submitted') +
  # xlab('') +
  # ylab('Number of variants (non-cumulative)\n') +
  xlab("") +
  ylab("") +
  ggtitle(expression(bold("Distribution of" ~ 1^st ~ "digit of pincode in India"))) +
  labs(#subtitle = "Estimation based on a multinomial fit to weekly genomic surveillance data deposited to GISAID",
        caption ="Source: https://www.geonames.org/export/zip")+#\uA9** Saket Choudhary") +
  theme(
    strip.text.x = element_text(size = 10, face = "bold", hjust = 0, color = "#225ea8"),
    strip.background = element_blank(),
    panel.grid.minor.x = element_blank(),
    panel.border = element_blank(),
    axis.title.x = element_text(size = 10),
    axis.title.y = element_text(size = 10),
    axis.text.x = element_text(angle = 90)
  ) +
  #scale_x_discrete(guide = guide_axis(angle = 30)) +
  theme(axis.text = element_text(size = 10, color = "black")) +
  #scale_fill_manual(values = rev(ggpubr::get_palette("Set2", k = 7)), name = "Pangolin lineage") +
  #scale_fill_manual(values=voc_colors, name="Pangolin lineage") +
guides(colour = guide_legend(override.aes = list(size=4))) +
dark_theme_gray() +
dark_theme_void()+
theme(
    plot.title = element_text(face = "bold"), # , hjust = 0.5),
    #plot.caption = element_markdown(hjust = 0),
axis.line=element_blank(),axis.text.x=element_blank(),
          axis.text.y=element_blank(),axis.ticks=element_blank(),
          axis.title.x=element_blank(),
    panel.grid.major = element_blank(), 
    panel.grid.minor =  element_blank(), 
          axis.title.y=element_blank(), legend.position="right", legend.direction="vertical"  )
```
